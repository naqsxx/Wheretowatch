```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_movies = pd.read_csv("IMDb movies.csv")

df_movies.head(2)
```

```python
df_movies = df_movies.dropna()
```

```python
df_movies["genre_1"] = df_movies["genre"].str.split(",").str.get(0)
df_movies["genre_2"] = df_movies["genre"].str.split(",").str.get(1)
df_movies["genre_2"].fillna(df_movies["genre_1"])
```

```python
df_movies[["genre","genre_1","genre_2"]]
```

```python
#Does the same thing as above but by using the apply function.

def get_first_genre(row):
    return row.split(",")[0]

def get_second_genre(row):
    genres = row.split(",")
    try:
        genre=genres[1]
    except IndexError:
        genre=genres[0]
    return genre
```

```python
df_movies["genre_1"] = df_movies["genre"].apply(get_first_genre)
df_movies["genre_2"] = df_movies["genre"].apply(get_second_genre)
```

```python
#Decide on what you want? Maybe only movies in English?

def english_language(row):
    return True if "English" in row else False
```

```python
df_movies["english"] = df_movies["language"].apply(english_language)
```

```python
print( "English:" + str(sum(df_movies["english"])) + "/ Not English: " + str(len(df_movies["english"]) - sum(df_movies["english"])))
```

```python
df_ratings = pd.read_csv("IMDb ratings.csv")
```

```python
df_ratings.dropna()
df_joined = pd.merge(df_movies, df_ratings, on="imdb_title_id")
df_joined.head(5)
```
